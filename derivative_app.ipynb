```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x=sp.symbols('x')
f=x**3-6*x**2 + 9*x + 15
```

```python
sp.pretty_print(f)
```

```python
f_lambdified=sp.lambdify(x,f,'numpy')
x_vals=np.linspace(-1,5,400)
y_vals=f_lambdified(x_vals)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=x**3 - 6*x**2 + 9x+15$', color='blue')

plt.title('Funcion original')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
f_prime=sp.diff(f,x)
f_prime_lambdified=sp.lambdify(x,f_prime,'numpy')
```

```python
sp.pretty_print(f_prime)
```

```python
y_prime_vals=f_prime_lambdified(x_vals)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=x**3 - 6*x**2 + 9x+15$', color='blue')
plt.plot(x_vals,y_prime_vals,label=r'Primera derivada ', color='red',linestyle='--')
plt.title('Funcion y su primera derivada')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
critical_points=sp.solve(f_prime)
print(critical_points)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=x**3 - 6*x**2 + 9x+15$', color='blue')
plt.plot(x_vals,y_prime_vals,label=r'Primera derivada ', color='red',linestyle='--')

plt.title('Funcion primera derivada y puntos criticos')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.scatter(critical_points, [f_lambdified(cp) for cp in critical_points], color='purple', label='puntos_criticos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
f_double_prime=sp.diff(f_prime,x)
f_double_prime_lambdified=sp.lambdify(x,f_double_prime,'numpy')
```

```python
y_double_prime_vals=f_double_prime_lambdified(x_vals)
```

```python
inflection_points=sp.solve(f_double_prime,x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals,y_vals,label=r'$f(x)=x**3 - 6*x**2 + 9x+15$', color='blue')
plt.plot(x_vals,y_prime_vals,label=r'Primera derivada ', color='red',linestyle='--')
plt.plot(x_vals,y_double_prime_vals,label=r'Segunda derivada', color='green',linestyle='-.')
plt.title('Funcion primera derivada y segunda derivada')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.scatter(critical_points, [f_lambdified(cp) for cp in critical_points], color='purple', label='puntos_criticos')
plt.scatter(inflection_points, [f_lambdified(ip) for ip in inflection_points], color='orange', label='puntos de infleccion')
plt.legend()
plt.grid(True)
plt.show()
```
